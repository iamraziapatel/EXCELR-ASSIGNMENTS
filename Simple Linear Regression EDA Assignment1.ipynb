```python
# Data loading:
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df_Delivery_data=pd.read_csv('delivery_time.csv')
# df_Delivery_data
```

```python
# Data cleaning: Checking and handling missing values, outliers,Check for duplicates and inconsistencies in the data.

#missing values check
missing_values_count =df_Delivery_data.isnull().sum()
print("Null values :\n",missing_values_count) 
missing_values_count.info()

#detecting outliers
sns.boxplot(data=df_Delivery_data)            
plt.show()

#duplicate check
duplicates= df_Delivery_data.duplicated()
print(duplicates.value_counts())
```

#using descriptive statistics to check inconsistency in the data.
Descriptive statistics like mean, median, mode, and standard deviation can reveal potential inconsistencies in data. 
If the mean and median are significantly different, this could be an indication of outliers or skewed data.
Similarly, if the standard deviation is high, this could be an indication of data points that are far from the average.

```python
# statistical Analysis

df_Delivery_data.describe()
df_Delivery_data.skew()

# statistical Analysis using visualizations
sns.distplot(df_Delivery_data['Delivery Time'],kde=True)
plt.show()
sns.distplot(df_Delivery_data['Sorting Time'],kde=True)
plt.show()
```

Insights : mean and Median are nearly equal so this indicates neither outliers and nor skewed data.
Hence after confirming from the dist plot as well that we can assume it to be normal distributed.
The data is clean.
