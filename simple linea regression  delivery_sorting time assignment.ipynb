```python
# EDA FOR cleaning ,imputation - null values replace with mean and median ,missing value handling for delivery_data
import pandas as pd
df_delivery = pd.read_csv('delivery_time.csv')
df_delivery.shape # check the size and shape

df_delivery.duplicated() #check for duplicate rows
df_delivery.drop_duplicates() # just in case drop if any duplicate rows , no duplicate columns exist
df_delivery.isnull() # check for null

# renaming the sorting Time column since it is giving error in linear regression Model
df_delivery.rename(columns={'Sorting Time':'Sorting_Time'},inplace=True)
```

```python
#Build a simple linear regression model
from scipy.stats import stats
x=df_delivery["Sorting_Time"].values
y=df_delivery["Delivery Time"].values
#print(x)
slope, intercept, r, p, std_err = stats.linregress(x, y) 
# checking reg intercept,slope ,r and p values both by pandas and skylearn models in y=mx+c where m is slope and c is intercept and r is correlation
print('intercept(c)=',intercept)
print('value of r=',r)
print('value of slope(m)=',slope)
#print(r)# relationship between the two features

# my function for predicting the future values
def myfunc(x):
  return slope * x + intercept
# The list() function creates a list object. A list object is a collection which is ordered and changeable. 
mymodel = list(map(myfunc, x)) #map() function returns a map object(which is an iterator) of the results after applying the given function to each item of a given iterable (list, tuple etc.) Syntax :map(fun, iter)
plt.scatter(x, y,marker='+')
plt.plot(x, mymodel)
plt.show()

print( 'prediction from custom made model from scipy package for sorting time 4 =' ,myfunc(4)) # predict the future values
```

```python
#select the best model using R or Python from SKlearn

from sklearn import linear_model
reg=linear_model.LinearRegression()

reg.fit(df_delivery[['Sorting_Time']],df_delivery['Delivery Time'])

# checking reg intercept,slope ,r and p values both by pandas and skylearn models in y=mx+c where m is slope and c is intercept and r is correlation
print('intercept=',reg.intercept_)
print('value of slope m=',reg.coef_)

# predicting values currently for hardcoded data
print("Prediction of delivery time from the sklearn package  for sorting time 4 = ",reg.predict([[4]]))
#plotting the regression using sklearn reg predict
plt.scatter(x=df_delivery['Sorting_Time'], y=df_delivery['Delivery Time'],color='red',marker='+')
plt.plot(df_delivery['Sorting_Time'],reg.predict(df_delivery[['Sorting_Time']]),color='violet')
plt.show() 
```

```python
#Transforming the data and Analysing the transformation and original distributions
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np


figure,axes = plt.subplots(1,2,sharex=True,figsize=(13,8))

figure.suptitle('COMPARING :THE Normal distribution before and after the log transformation on DELIVERY TIME')
axes[0].set_title('Delivery Time PDF')



#transforming the data and checking for the normal curve
axes[1].set_title('log transformed PDF of Delivery Time')
sns.distplot(df_delivery['Delivery Time'],color="r",ax=axes[0],label="Orginal Skew:{0}".format(np.round(df_delivery['Delivery Time'].skew(),4)))

log_DeliveryTime =np.log1p(df_delivery['Delivery Time'])
log_DeliveryTime=pd.DataFrame(log_DeliveryTime)
log_DeliveryTime.rename({'Delivery Time':'Log_DeliveryTime'})

sns.distplot(log_DeliveryTime,label='after Log - skew',color='green',ax=axes[1])
plt.show()

print('****Analysis of the original and Transformed Data*****')
print("original Skew :",df_delivery['Delivery Time'].skew()," \nlog transform Skew:",log_DeliveryTime['Delivery Time'].skew())
print('=====original Data statistics ===== \n',df_delivery['Delivery Time'].describe())
print('====log transformation data statistics====\n',log_DeliveryTime.describe())


# TRansformation of Sorting Time 

figure1,axes1 = plt.subplots(1,2,sharex=True,figsize=(13,8))

figure1.suptitle('COMPARING :THE Normal distribution before and after the log transformation on Sorting TIME')
axes1[0].set_title('Sorting Time PDF')



#transforming the data and checking for the normal curve
axes1[1].set_title('log transformed PDF of Sorting Time')
sns.distplot(df_delivery['Sorting_Time'],color="r",ax=axes1[0])

log_SortingTime =np.log1p(df_delivery['Sorting_Time'])
log_SortingTime=pd.DataFrame(log_SortingTime)
log_SortingTime.rename({'Delivery Time':'Log_DeliveryTime'})

# square transformation
sq_DeliveryTime=df_delivery['Delivery Time']**2
sq_DeliveryTime= pd.DataFrame(sq_DeliveryTime)
sq_SortingTime =df_delivery['Sorting_Time']**2
sq_SortingTime= pd.DataFrame(sq_SortingTime)

#
# square ROOT transformation
sqRT_DeliveryTime=df_delivery['Delivery Time']**(1/2)
sqRT_DeliveryTime= pd.DataFrame(sqRT_DeliveryTime)
sqRT_SortingTime =df_delivery['Sorting_Time']**(1/2)
sqRT_SortingTime= pd.DataFrame(sqRT_SortingTime)

sns.distplot(log_DeliveryTime,label='after Log - skew',color='green',ax=axes1[1])
plt.show()
print('****Analysis of the original and Transformed Data*****')
print("original Skew :",df_delivery['Sorting_Time'].skew()," \nlog transform Skew:",log_DeliveryTime['Delivery Time'].skew())
print('=====original Data statistics ===== \n',df_delivery['Sorting_Time'].describe())
print('====log transformation data statistics====\n',log_DeliveryTime.describe())

```

```python
# performance of prediction model after log transformation of both the input and response -RMSE Check for SLR models over LOG transformed data
from sklearn.metrics import mean_squared_error
from scipy.stats import stats
import math

# Performance check of the model by RMSE
originalmse = mean_squared_error(x, mymodel)
originalrmse = math.sqrt(originalmse)
print(originalrmse)

x=log_SortingTime['Sorting_Time'].values
y=log_DeliveryTime['Delivery Time'].values
#print(x)

slope, intercept, r, p, std_err = stats.linregress(x, y) 
# checking reg intercept,slope ,r and p values both by pandas and skylearn models in y=mx+c where m is slope and c is intercept and r is correlation
print('intercept(c)=',intercept)
print('value of r=',r)
print('value of slope(m)=',slope)
#print(r)# relationship between the two features

# my function for predicting the future values
def myfunc(x):
  return slope * x + intercept
# The list() function creates a list object. A list object is a collection which is ordered and changeable. 
mymodelpredic_tvalue = list(map(myfunc, x)) #map() function returns a map object(which is an iterator) of the results after applying the given function to each item of a given iterable (list, tuple etc.) Syntax :map(fun, iter)
plt.scatter(x, y,marker='+')
plt.plot(x, mymodelpredic_tvalue)
plt.show()
true_val=x
print( 'prediction from custom made model from scipy package for log transformed with sorting time 4 =' ,myfunc(4)) # predict the future values
mse = mean_squared_error(true_val, mymodelpredic_tvalue)
rmse = math.sqrt(mse)
# square TRANSFORMATION
sqx=sq_SortingTime['Sorting_Time'].values
sqy=sq_DeliveryTime['Delivery Time'].values
slopeSQ, interceptSQ, rSQ, pSQ, std_errSQ = stats.linregress(sqx, sqy) 
def myfuncSQ(x):
  return slopeSQ *x + interceptSQ

mymodelpredic_SQvalue = list(map(myfuncSQ, sqx))
SQmse = mean_squared_error(sqy, mymodelpredic_SQvalue)
SQrmse = math.sqrt(SQmse)

# SQROOT TRANSFORMATION
sqRTx=sqRT_SortingTime['Sorting_Time'].values
sqRTy=sqRT_DeliveryTime['Delivery Time'].values
slopeRT, interceptRT, rRT, pRT, std_errRT = stats.linregress(sqRTx, sqRTy) 
def myfuncSQ(x):    
  return slopeRT *x + interceptRT

mymodelpredic_SQRTvalue = list(map(myfunc, sqRTx))
SQRTmse = mean_squared_error(sqRTy, mymodelpredic_SQRTvalue)
SQRTrmse = math.sqrt(SQRTmse)

print("Mean Squared error for original data = ",originalmse)
print("Mean Squared error for log transformed data = ",mse)
print("Mean Squared error for SQUARE transformed data = ",SQmse)
print("Mean Squared error for SQUAREROOT transformed data = ",SQRTmse)
print("root Mean Squared error for original data = ",originalrmse)
print("root Mean Squared error for log normal data = ",rmse)
print("root Mean Squared error for SQUARE normal data = ",SQrmse)
print("root Mean Squared error for SQUARE normal data = ",SQRTrmse)
```
